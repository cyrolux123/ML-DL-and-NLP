```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## Create a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

## Create a customized line plot
plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2, markersize=10) 
plt.grid(True)
```
